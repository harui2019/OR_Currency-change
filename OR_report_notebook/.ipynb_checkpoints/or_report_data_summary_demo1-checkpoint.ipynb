```python
plt.figure(figsize = (20, 10)).patch.set_facecolor('white')
plt.plot(rate_1['rate_1(k)'])
plt.plot([0,1000],[1,1], color ='r')
plt.xticks(range(9, 1000 ,50))
plt.ylabel("remain rate")
```

```python
rate_1['(1-remain rate)/remain rate'] = (1 - rate_1['rate_1(k)']/rate_1['rate_1(k)'])
```

```python
plt.figure(figsize = (20, 5)).patch.set_facecolor('white')
plt.plot(rate_1['(1-remain rate)/remain rate'])
plt.plot([0,1000],[1,1], color ='r')
plt.xticks(range(9, 1000 ,50))
plt.ylabel("(1-remain rate)/remain rate")
```
